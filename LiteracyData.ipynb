```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats


literacy = pd.read_csv('Literacy rates (no pw2).csv')
literacy.rename(columns={'Literacy rate': 'LiteracyRate'},inplace=True)
print(literacy.head(5))
print(literacy.info())
print(literacy.describe())
```

# Which country has highest literacy

```python
max_lit = literacy.groupby('LiteracyRate')['Country'].max()
print(max_lit)
```

```python
sns.boxplot(data=literacy, x='LiteracyRate',y='Region').set(title = 'Literacy Rate per World Region')
plt.show()
plt.close()
# So this averages lit rate across all years.
#Then we want max and min year- so then we can break this graph by year, like we did below
#Then we can do by year and by gender 2x2 type table
# Further we can maybe break it down by countries in each region? BY YEAR? BY GENDER? Or is this data dredging?
```

# Boxplot with x labels at 45

```python
ax= sns.boxplot(data=literacy, x='Region',y='LiteracyRate')
plt.setp(ax.get_xticklabels(), rotation=45)
plt.show()
plt.close()
```

# Over what period has the data been recorded?

```python
print(literacy.Year.min())
print(literacy.Year.max())
```

```python
lit_2010= literacy[literacy.Year == 2010]
ax = sns.boxplot(data=lit_2010, x= 'LiteracyRate', y='Region').set(title = 'Literacy Rate per World Region in 2010')
sns.set(rc={'figure.figsize':(8,8)})
plt.show()
plt.close()
```

```python
lit_2018= literacy[literacy.Year == 2018]
ax = sns.boxplot(data=lit_2018, x= 'LiteracyRate', y='Region').set(title = 'Literacy Rate per World Region in 2018')
sns.set(rc={'figure.figsize':(8,8)})
plt.show()
plt.close()
```

```python
literacy.loc[literacy["Gender"] == "male", "Gender"] = 1
literacy.loc[literacy["Gender"] == "female", "Gender"] = 2
literacy = literacy.drop(literacy[literacy['Gender'] == 'total'].index)
```

# Quick look at the literacy rates per gender

```python
sns.boxplot(data = literacy,x='Gender',y='LiteracyRate').set(title = 'Literacy Rate per Gender')
plt.show()
plt.close()
```
